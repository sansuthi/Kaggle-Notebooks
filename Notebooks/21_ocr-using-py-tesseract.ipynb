## Importing Libraries

```python
from PIL import Image
import pytesseract
```

## Loading & displaying image

```python
image = Image.open('../input/dataset-scanned-documents-ocr/scannedpaper1-jpeg.jpg')
display(image)
```

## Size of Image

```python
image.size
```

# Image to Text

```python
text = pytesseract.image_to_string(image)
print(text)
```

## Cropping image & extracting text

```python
# FORMULA 1
bounding_box = (490,470,680,490)
formula1 = image.crop(bounding_box)
display(formula1)
```

```python
formula1.size
```

```python
# Resizing the extracted image
size = (formula1.width*2,formula1.height*2)
styles = [Image.HAMMING, Image.BICUBIC, Image.NEAREST, Image.BOX, Image.BILINEAR, Image.LANCZOS]
labels = ['HAMMING', 'BICUBIC', 'NEAREST', 'BOX', 'BILINEAR', 'LANCZOS']
for i in range(len(styles)):
    print(labels[i])
    display(formula1.resize(size, styles[i]))
```

```python
# FORMULA 2
bounding_box = (440,560,690,580)
formula2 = image.crop(bounding_box)
display(formula2)
```

```python
formula2.size
```

```python
size = (formula2.width*2,formula1.height*2)
display(formula2.resize(size, Image.NEAREST))
```

```python
# FORMULA 3
bounding_box = (510,640,660,660)
formula3 = image.crop(bounding_box)
display(formula3)
```

## Drawing Rectangles

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
fig, ax = plt.subplots(figsize=(15,20))
ax.imshow(image)
rect = patches.Rectangle((75,98),700,100, linewidth=3, edgecolor='b', facecolor='none')
ax.add_patch(rect)

rect = patches.Rectangle((485,468),200,25, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)
rect = patches.Rectangle((445,558),245,25, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)
rect = patches.Rectangle((510,635),150,25, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)

rect = patches.Rectangle((70,205),710,70, linewidth=1.5, edgecolor='g', facecolor='none')
ax.add_patch(rect)

rect = patches.Rectangle((750,1130),30,30, linewidth=2, edgecolor='purple', facecolor='none')
ax.add_patch(rect)

plt.show()
```
