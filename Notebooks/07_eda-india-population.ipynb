```python
# importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading Data
data = pd.read_csv('../input/indian-population/IndiaPopulation_2021.csv')
data.head()
```

```python
# Sorting Dataframe: 1950 - 2021
data = data.sort_values('Year').set_index('Year')
data
```

```python
# BAR PLOT
plt.figure(figsize=(15,5))
plt.title('Barplot for Population by Year', fontsize=15)
sns.barplot(data.index, data.Population)
plt.show()
```

```python
# HIST PLOT
plt.figure(figsize=(10,5))
sns.histplot(data)
plt.show()
```

```python
data.plot()
plt.show()
```

```python
# PAIR PLOT
sns.pairplot(data, kind='reg')
plt.show()
```

```python
# SCATTER PLOT
plt.figure(figsize=(10,5))
plt.title('Scatterplot for Population')
sns.scatterplot(data.index, data.Population,
                s=100, alpha=0.5, color='red')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Scatterplot for GrowthRate')
sns.scatterplot(data.index, data.GrowthRate,
                s=100, alpha=0.5, color='green')
plt.show()
```

```python
# Normalizing Data
norm_data = (data - data.min())/(data.max()-data.min())
norm_data
```

```python
norm_data.plot()
plt.show()
```

```python
# HEAT MAP - PLOTTING CORRELATION
sns.heatmap(norm_data.corr(), annot=True)
plt.show()
```

## Year-wise Analysis

```python
year_data = data.iloc[0:len(data):10]
year_data
```

```python
# POINT PLOT
plt.figure(figsize=(10,5))
sns.pointplot(year_data.index, year_data.Population, color='red')
plt.show()
```

```python
# BAR PLOT
plt.figure(figsize=(10,5))
sns.barplot(year_data.index, year_data.Population)
plt.show()
```

## Growth Rate Analysis

```python
plt.figure(figsize=(10,5))
plt.title('Scatter Plot - Growth Rate', fontsize=15)
sns.scatterplot(data.index, data.Population, 
                hue=data.GrowthRate,s=75)
plt.show()
```

```python
plt.title('Boxplot for Growth Rate', fontsize=15)
sns.boxplot(y=data.GrowthRate, color='red')
plt.show()
plt.title('Violinplot for Growth Rate', fontsize=15)
sns.violinplot(y=data.GrowthRate)
plt.show()
plt.title('Stripplot for Growth Rate', fontsize=15)
sns.stripplot(y=data.GrowthRate, color='green',
              s=10, alpha=0.7)
plt.show()
```

```python
gr_year = data[data['GrowthRate']>2]
gr_year
```

```python
plt.figure(figsize=(10,5))
plt.title('Scatter Plot - Growth Rate > 2', fontsize=15)
sns.scatterplot(gr_year.index, gr_year.Population, 
                hue=gr_year.GrowthRate,s=100)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Count Plot - Growth Rate > 2', fontsize=15)
sns.countplot(gr_year.GrowthRate)
plt.show()
```
